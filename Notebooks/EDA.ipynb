## Univariate Analysis

```python
sns.histplot(df["release_date"], kde=True)
plt.show()
```

```python
sns.histplot(df["len"] , kde=True)
plt.show()
```

The length of the songs shows a right skew, with a majority of songs between 25 and 100

```python
sns.histplot(df["genre"])
plt.show()
```

A vast majority of the songs seem to be either pop or country, with blues being a close second. The hip hop genre has the least amount of songs in this dataset

```python
sns.histplot(df["topic"])
plt.show()
```

We can observe that songs about feelings make up around 500 songs in this dataaset while the largest category seems to be sadness which has more than 6000 songs.

## Bivariate Analysis

Since most of the descriptive columns such as song name, artist name and genre are objects, I will first generate some scatter plots between the numeric columns to analyze the distributions of various scores in the dataset

```python
sns.scatterplot(data=df, x="dating", y="violence")

plt.show()
```

```python
sns.scatterplot(data=df, x="feelings", y="sadness")

plt.show()
```

```python
sns.scatterplot(data=df, x="obscene", y="music")

plt.show()
```

```python
#I wanted to observe whether the genre has an impact on the length of the song
sns.boxplot(data=df, x="genre", y="len")
plt.show()
```

```python
sns.boxplot(data=df, x="topic", y="len")
plt.show()
```

```python
sns.boxplot(data=df, x="genre", y="movement/places")
plt.show()
```

```python
#Creating a df with only numeric columns for correlation matrix
Numeric = df.select_dtypes(include = "number")
```

```python
#Correlation matrix of all features

Correlation_Matrix = Numeric.corr()
plt.figure(figsize=(15,10))
sns.heatmap(Correlation_Matrix, annot = True )
plt.title("Correlation Matrix")
plt.show()
```
