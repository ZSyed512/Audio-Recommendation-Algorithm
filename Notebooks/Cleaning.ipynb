```python
import pandas as pd
import numpy as np
```

```python
#Since the test set also needs to be in the same format, i will transformm both sets together
df = pd.read_csv("../data/train (1).csv")
df_Test = pd.read_csv("../data/recommend.csv")
```

```python
#Checking for null values
df.isna().sum()
```

```python
df.head()
```

```python
#Dropping the lyrics column
#Since the test set, rcommend.csv has to be in the same format, i wil preprocess both sets together
df.drop("lyrics" , axis = 1 , inplace = True)
df_Test.drop("lyrics" , axis = 1 , inplace = True)
```

```python
#Dropping Unnamed 
#Dropping various columns that would not be helpful for my approach, which is to try and generate clusters
#based on descriptive characteristics about the songs content, genre rather than specific length or song information.

df = df.drop(["Unnamed: 0" ,  "artist_name" , "track_name" , "release_date" , "len" , "age"] , axis = 1)
df_Test = df_Test.drop(["Unnamed: 0" ,  "artist_name" , "track_name" , "release_date" , "len" , "age"] , axis = 1)
```

```python
df.head()
```

```python
#Generating one-hot-encoding for the topic and genre columns using pd.get_dummies

#Temporary one hot encodings with prefix genre
Temp_OHE_Genre = pd.get_dummies(df["genre"] , prefix = "genre").astype(int)

#Final dataframe, joining one hot encodings
df_final_Genre = df.join(Temp_OHE_Genre)

df_final_Genre.drop("genre" , axis = 1 , inplace = True)

df_final_Genre.head()


#For test set
#Generating one-hot-encoding for the topic and genre columns using pd.get_dummies

#Temporary one hot encodings with prefix genre
Temp_OHE_Genre_Test = pd.get_dummies(df_Test["genre"] , prefix = "genre").astype(int)

#Final dataframe, joining one hot encodings
df_final_Genre_Test = df_Test.join(Temp_OHE_Genre_Test)

df_final_Genre_Test.drop("genre" , axis = 1 , inplace = True)



```

```python
#Temporary one hot encodings with prefix topic
Temp_OHE_topic = pd.get_dummies(df_final_Genre["topic"] , prefix = "topic").astype(int)

#Final dataframe, joining one hot encodings
df_final = df_final_Genre.join(Temp_OHE_topic)

df_final.drop("topic" , axis = 1 , inplace = True)

df_final.head()


#For test set
#Temporary one hot encodings with prefix topic
Temp_OHE_topic_Test = pd.get_dummies(df_final_Genre_Test["topic"] , prefix = "topic").astype(int)

#Final dataframe, joining one hot encodings
df_final_Test = df_final_Genre_Test.join(Temp_OHE_topic_Test)

df_final_Test.drop("topic" , axis = 1 , inplace = True)
```

```python
df_final.isna().sum()
```

```python
df_final.to_csv("../data/Cleaned.csv",index=False)
df_final_Test.to_csv("../data/TestSetCleaned.csv",index=False)
```
